```python
# Extract Hour
df['Hour'] = df['Time'].apply(lambda x: x.hour)
df['Hour'].head()
```

```python
# Create time slot
def time_slot(x):
    if 6 <= x <= 11:
        return 'Morning'
    elif 12 <= x <= 16:
        return 'Afternoon'
    elif 17 <= x <= 21:
        return 'Evening'
    else:
        return 'Night'

df['time_slot'] = df['Hour'].apply(time_slot)
```

```python
df.groupby('time_slot').size()
```
